What is the average step count on Sad mood days?

```python
# step_count[mood == 'Sad'].astype(float).mean()
# round(step_count[mood == 'Sad'].astype(float).mean(), 2)
# round(float(step_count[mood == 'Sad'].mean()),2)
np.mean(step_count[mood == 'Sad'])
```

What is the average step count on Happy mood days?

```python
# round(float(step_count[mood=='Happy'].mean()),2)
np.mean(step_count[mood == 'Happy'])
```

Is there a relation between mood and step count?
Yes(More step count more Happy)

```python
unique_moods = np.unique(mood)

for m in unique_moods:
    avg_steps = np.mean(step_count[mood == m])
    print(f"Average step count for mood {m}: {avg_steps}")
```

What is the mood when step count was greater than 4000?

```python
np.unique(mood[step_count>4000],return_counts = True)
```

Out of 38 days when step count was more than 4000, user was feeling happy on 22 days.

What is the mood when step count was less than or equal to 2000?

```python
np.unique(mood[step_count<=2000],return_counts=True)
```

Observation:
The user felt Happy on 13 days.

The user felt Neutral on 8 days.

The user felt Sad for on 18 days.

Conclusion:
The mood Sad was recorded most frequently.

The user experienced Happy mood less often than Sad.

This suggests that the user had more days with negative or low moods than positive moods during the recorded period.
